## CODE TO PERFORM ROTATION, TRANSLATION AND RESIZING


# LOAD/DISPLAY IMAGE

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline

image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/section 5 - Computer Vision/Self_Driving_Car.jpg')
cv2_imshow(image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
height, width = image.shape[:2]
```

```python
height
```

```python
width
```

# PERFORM ROTATIONS

```python
M_rotation = cv2.getRotationMatrix2D((width/2, height/2), 90, 0.5) # Rotate around the center
# Center, angle of rotation, rescaling
rotated_image = cv2.warpAffine(image, M_rotation, (width, height)) # insert size of the output image

# cv2.imshow('Rotated', rotated_image)
cv2_imshow(rotated_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

# PERFORM TRANSLATIONS

```python
Translational_Matrix = np.float32([[1, 0, 300], # y
                                  [0, 1, 50]]) # x
```

```python
translated_image = cv2.warpAffine(image, Translational_Matrix, (width, height))
# cv2.imshow('Translated Image', translated_image)
cv2_imshow(translated_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Don't worry about the color, recall openCV uses BGR instead of RGB
plt.imshow(translated_image)
```

# PERFORM RESIZING

```python
resized_image = cv2.resize(image, None, fx=4, fy=4, interpolation = cv2.INTER_CUBIC) # Try 0.5
# cv2.imshow('Resized Image', resized_image)
cv2_imshow(resized_image)  
cv2.waitKey()
cv2.destroyAllWindows()
```
