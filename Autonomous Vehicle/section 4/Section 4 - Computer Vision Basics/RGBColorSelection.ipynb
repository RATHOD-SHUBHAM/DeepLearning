# Code to perform Color Selection on colored image
Select one perticular color from an image.

Eg. if we want to extract white color. Then extract the pixels that are close to 255.

# Step 1 : Import all the necessary libraries

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

# Step 2 : Read and display image

```python
imageColor = mpimg.imread('/content/drive/MyDrive/Colab Notebooks/Section 4 - Computer Vision Basics/Copy of image_lane_c.jpg')
imageColor.shape
```

# Step 3 : Extract the required pixel from image.
Here we will extract color white. pixel value nearly equal to 200

```python
# making a copy of actual image
imageCopy = np.copy(imageColor)
imageCopy.shape
```

Unlike gray image RGB img has 3 layer 

```python
imageCopy[ (imageCopy[: , : , 0] < 200) | (imageCopy[: , : , 1] < 200) | (imageCopy[: , : , 2] < 200) ] = 0
```

```python
plt.imshow(imageCopy, cmap = 'gray')
plt.show()
```

```python
cv2.imwrite("/content/drive/MyDrive/Colab Notebooks/Section 4 - Computer Vision Basics/Colorwhite.jpg",imageCopy)
```
