#  CODE FOR IMAGE MANIPULATIONS (Edge Detection using Sobel/Laplacian and Canny) 

# Import all the necessary libraries

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np

import cv2
```

# Read the image and display it using OpenCV


```python
from google.colab.patches import cv2_imshow

image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Section 4 - Computer Vision Basics/Copy of Self_Driving_Car.jpg')

cv2_imshow(image)

# 'waitKey' allows us to input information when a image window is open
# By leaving it blank it just waits for anykey to be pressed before continuing. 
cv2.waitKey()
cv2.destroyAllWindows()
```

# Let's print out the dimensions of convert to grayscale image

```python
image.shape
```

```python
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

# Let's Get Sobel X and Y

```python
# Sobel Calculations
x_sobel = cv2.Sobel(gray_img, cv2.CV_64F, 0, 1, ksize = 7) # x = 0 ,1
cv2_imshow(x_sobel)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
y_sobel = cv2.Sobel(gray_img, cv2.CV_64F, 1, 0, ksize = 7) # y = 1, 0
cv2_imshow( y_sobel)
cv2.waitKey()
cv2.destroyAllWindows()
```

# Let's get Laplacian 

```python
laplacian = cv2.Laplacian(gray_img, cv2.CV_64F)
cv2_imshow(laplacian)
cv2.waitKey()
cv2.destroyAllWindows()

```

# Let's get Canny 

```python
# threshold1 – first threshold for the hysteresis procedure.
# threshold2 – second threshold for the hysteresis procedure.
# Let's get Canny Edge

threshold_1 = 120
threshold_2 = 200


canny = cv2.Canny(gray_img, threshold_1, threshold_2)

cv2_imshow(canny)
cv2.waitKey()
cv2.destroyAllWindows()
```
