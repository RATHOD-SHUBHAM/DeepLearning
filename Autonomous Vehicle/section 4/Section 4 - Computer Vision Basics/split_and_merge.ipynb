```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

# Split Image

```python
image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Section 4 - Computer Vision Basics/Copy of Self_Driving_Car.jpg')
B, G, R = cv2.split(image)
B.shape
G.shape
R.shape
```

```python
# Let's view the R, B, G channels
cv2_imshow(B) # this will be read as a single channel

# since it is a single channel. We will get a grey image.
```

```python
# We create our own 3D image for the blue channel!
zeros = np.zeros(image.shape[:2], dtype = "uint8")
# add zero for the remaining 2 channel, then merge the 3 channel keeping blue as the first channel
cv2_imshow(cv2.merge([B, zeros, zeros]))
```

```python
# Let's merge RGB channels to create our original image
image_merged = cv2.merge([B, G, R]) 
cv2_imshow(image_merged)
```

```python
#We can make it more green
image_merged = cv2.merge([B, G+100, R]) 
cv2_imshow(image_merged)
```
