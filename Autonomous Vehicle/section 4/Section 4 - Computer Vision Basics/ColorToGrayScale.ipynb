# Code to convert color image to gray scale

# Step 1 : Import all the necessary libraries

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

 Read and display image

```python
imageColor = mpimg.imread('/content/drive/MyDrive/Colab Notebooks/Section 4 - Computer Vision Basics/Copy of image_lane_c.jpg')
```

```python
plt.imshow(imageColor)
```

Display the shape - That is how many pixels as heigh and width and also the layer.

For colored image there should be 3 layer

```python
imageColor.shape
```

# Step 2 : Convert image to grayscale


```python
imageGray = cv2.cvtColor(imageColor , cv2.COLOR_RGB2GRAY)
plt.imshow(imageGray , cmap = 'gray') # without cmap. we will get something like hue image
```

```python
# color image will have 3 layer and gray scale should not habe any layer
imageGray.shape # this should not show any layer.
```

```python
# imageGray is a array of pixels
```

```python
imageGray # this will show the value of pixels
# 0 - 255 , from black top white
```
