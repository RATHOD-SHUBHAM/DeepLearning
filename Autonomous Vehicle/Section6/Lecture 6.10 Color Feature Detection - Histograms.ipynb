## CODE TO PERFORM HISTOGRAM OF COLOR


# LOAD/DISPLAY IMAGE

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
image_color = cv2.imread('/content/drive/MyDrive/Self Driving Car/SelfDrivingMaterials/Section6/tesla.jpg')

# cv2.imshow('Original Image', image_color)
cv2_imshow(image_color)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(image_color)
```

```python
image_color.shape
```

```python
color = ('b', 'g', 'r')

for i, col in enumerate(color):
    histogram2 = cv2.calcHist([image_color], [i], None, [256], [0, 256])
    plt.plot(histogram2, color = col)
    plt.xlim([0,256])
    
plt.show()
```
