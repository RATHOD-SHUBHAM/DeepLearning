## CODE TO PERFORM IMAGE PYRAMIDING

# LOAD/DISPLAY IMAGE

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
image_color = cv2.imread('/content/drive/MyDrive/Self Driving Car/SelfDrivingMaterials/Section6/Self_Driving.jpg')
# image_color = cv2.imread('truck_back_color.jpg')

# cv2.imshow('Original Image', image_color)
cv2_imshow(image_color)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(image_color)
```

```python
image_color.shape
```

# PERFROM PYRAMIDING (DOWN)

```python
scaled_down_1 = cv2.pyrDown(image_color)
scaled_down_1.shape
```

```python
scaled_down_2 = cv2.pyrDown(scaled_down_1)
scaled_down_2.shape
```

```python
scaled_down_3 = cv2.pyrDown(scaled_down_2)
scaled_down_3.shape
```

```python
# cv2.imshow('Original', image_color)
# cv2.imshow('Smaller #1', scaled_down_1)
# cv2.imshow('Smaller #2', scaled_down_2)
# cv2.imshow('Smaller #3', scaled_down_3)


cv2_imshow( image_color)
cv2_imshow(scaled_down_1)
cv2_imshow(scaled_down_2)
cv2_imshow( scaled_down_3)

cv2.waitKey()
cv2.destroyAllWindows()
```


# PERFROM PYRAMIDING (UP)

```python
scaled_up = cv2.pyrUp(image_color)
scaled_up.shape
```

```python
# cv2.imshow('Larger ', scaled_up)
cv2_imshow(scaled_up)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
